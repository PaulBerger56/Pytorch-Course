## Introduction to Tensors

### Creating Tensors

PyTorch tensors are created using `torch.Tensor()`

```python
# scalar
scalar = torch.tensor(7)
scalar
```

```python
# ndim = number of dimensions
scalar.ndim
```

```python
# Get tensor back as Python int
scalar.item()
```

```python
# Vector
vector = torch.tensor([7,7])
vector
```

```python
vector.ndim
```

```python
vector.shape
```

```python
# MATRIX
MATRIX = torch.tensor([[7,8],
                       [9,10]])
MATRIX
```

```python
MATRIX.ndim
```

```python
MATRIX[1]
```

```python
# TENSOR
TENSOR = torch.tensor([[[1,2,3],
                        [3,6,9],
                        [2,5,4]]])
TENSOR
```

```python
TENSOR.ndim
```

```python
TENSOR.shape
```

```python
TENSOR[0]
```
